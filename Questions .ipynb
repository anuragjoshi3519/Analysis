# Q5. Compare the sex ratio of urban areas and rural areas using graph.

```python
d4=df  
d4=d4.drop(d4[d4['Total/Rural/Urban']=='Total'].index)
d4=d4.drop(d4[d4['Adolescent and youth']!='All Ages'].index)
d4=d4.reset_index(drop=True)
d4=d4.drop([i for i in range(12)])
d4=d4.reset_index(drop=True)
d4['Sex Ratio']=d4['Total Male']/d4['Total Female']
```

```python
d41=d4[d4['Total/Rural/Urban']=='Rural']                  #extracting Rural data
d42=d4[d4['Total/Rural/Urban']=='Urban']                  #extracting urban data
d43=d41[['Area Name','Total/Rural/Urban','Sex Ratio']]       #Rural Sex ratio Table
d44=d42[['Area Name','Total/Rural/Urban','Sex Ratio']]       #Urban Sex Ratio Table
```

```python
plt.barh(d43['Area Name'],d43['Sex Ratio'],alpha=0.3,color='r')
plt.barh(d44['Area Name'],d44['Sex Ratio'],alpha=0.3,color='g')
plt.xlabel('Rural and Urban Sex Ratio',fontsize=16)
l=['Red=Rural Sex Ratio','Green=Urban Sex Ratio']
plt.legend(l,loc=1,fontsize=15)
plt.title('Comparison between Rural and Urban Areas Sex Ratio',fontsize=18)
plt.show()
```

# Q6. Which state has the smallest and the largest difference in the sex ratios of its rural and urban areas?

```python
d41=d41.reset_index(drop=True)
d42=d42.reset_index(drop=True)
L=[ (d41['Area Name'][i],abs(d41['Sex Ratio'][i]-d42['Sex Ratio'][i])) for i in range(d41.shape[0])]
```

# Q11. Tell whether in youth ages the no. of marginal workers is more or the no. of main workers.

```python
d11=df  
d11=d11.drop(d11[d11['Total/Rural/Urban']!='Total'].index)
d11=d11.drop(d11[d11['Adolescent and youth']!='Youth (15-24)'].index)
d11=d11.reset_index(drop=True)
d11=d11.drop([i for i in range(12)])
d11=d11.reset_index(drop=True)
d11['TotalMarginal']=d11['MarginalWorker-M']+d11['MarginalWorker-F']
d11['TotalMainWorker']=d11['MainWorker-F']+d11['MainWorker-M']
TotalMarginal = d11['TotalMarginal'].sum()
TotalMain = d11['TotalMainWorker'].sum()
if(TotalMarginal>TotalMain):
    print("In youth ages the no. of marginal workers is more")
else:
    print("In youth ages the no. of main workers is more")
```

# Q13. Which state is home to the maximum no. of ST in India? Which state has the minimum no. of ST in India?


```python
d13=df  
d13=d13.drop(d13[d13['Total/Rural/Urban']!='Total'].index)
d13=d13.drop(d13[d13['Adolescent and youth']!='All Ages'].index)
d13=d13.reset_index(drop=True)
d13=d13.drop([i for i in range(12)])
d13=d13.reset_index(drop=True)
d13['Total ST']=d13['ST-M']+d13['ST-F']
d13['Total']=d13['Total Male']+d13['Total Female']

d13['Percent'] = (d13['Total ST']*100) / d13['Total']
d113=d13[['Area Name','Percent']]
print(d113[d113['Percent']==d113['Percent'].max()])
print(d113[d113['Percent']==d113['Percent'].min()])
```
