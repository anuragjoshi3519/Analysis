# Q5. Compare the sex ratio of urban areas and rural areas using graph.

```python
d4=df  
d4=d4.drop(d4[d4['Total/Rural/Urban']=='Total'].index)
d4=d4.drop(d4[d4['Adolescent and youth']!='All Ages'].index)
d4=d4.reset_index(drop=True)
d4=d4.drop([0,1])
d4=d4.reset_index(drop=True)
d4['Sex Ratio']=d4['Total Male']/d4['Total Female']
```

```python
d41=d4[d4['Total/Rural/Urban']=='Rural']                  #extracting Rural data
d42=d4[d4['Total/Rural/Urban']=='Urban']                  #extracting urban data
d43=d41[['Area Name','Total/Rural/Urban','Sex Ratio']]       #Rural Sex ratio Table
d44=d42[['Area Name','Total/Rural/Urban','Sex Ratio']]       #Urban Sex Ratio Table
```

```python
plt.rcParams['figure.figsize'] = [12, 12]
plt.barh(d43['Area Name'],d43['Sex Ratio'],alpha=0.3,color='r')
plt.barh(d44['Area Name'],d44['Sex Ratio'],alpha=0.3,color='g')
plt.xlabel('Rural and Urban Sex Ratio',fontsize=16)
l=['Red=Rural Sex Ratio','Green=Urban Sex Ratio']
plt.legend(l,loc=1,fontsize=15)
plt.title('Comparison between Rural and Urban Areas Sex Ratio',fontsize=18)
plt.show()
```

# Q6. Which state has the smallest and the largest difference in the sex ratios of its rural and urban areas?

```python
d41=d41.reset_index(drop=True)
d42=d42.reset_index(drop=True)
L=[ (d41['Area Name'][i],abs(d41['Sex Ratio'][i]-d42['Sex Ratio'][i])) for i in range(d41.shape[0])]
```

```python
L
```

```python
maxSR=max(L, key=lambda L: L[1])
minSR=min(L, key=lambda L: L[1])
print("State with the largest difference in Sex Ratio is {} with difference {}".format(maxSR[0],maxSR[1]))
print("State with the smallest difference in Sex Ratio is {} with difference {}".format(minSR[0],minSR[1]))
```

# Q9. Calculate and show the percentage of child labourers in each state using graph.(consider the category 10-14 age group)


```python
d9=df  
d9=d9.drop(d9[d9['Total/Rural/Urban']!='Total'].index)
d9=d9.drop(d9[d9['Adolescent and youth']!='10-14'].index)
d9=d9.reset_index(drop=True)
d9=d9.drop([0])
d9=d9.reset_index(drop=True)
d9['Total Children']=d9['Total Male']+d9['Total Female']
d9['Labourers']=d9['MainWorker-F']+d9['MainWorker-M']+d9['MarginalWorker-M']+d9['MarginalWorker-F']
d9['% of Child Labourers'] = (d9['Labourers']*100) / d9['Total Children']
d91=d9[['Area Name','Total Children','% of Child Labourers']]
d91
```

```python
plt.rcParams['figure.figsize'] = [12, 12]
plt.barh(d91['Area Name'],d91['% of Child Labourers'],alpha=0.3,color='r')
plt.title('Percentage of Child Labourers (Age 10-14) in each state.',fontsize=16)
plt.xlabel('Percentage',fontsize=16)
plt.show()
```

# Q10. Calculate the literacy rate of people belonging to scheduled castes in rural areas of each state. In which state the literacy rate of SC the least?

```python
d10=df
d10=d10.drop(d10[d10['Total/Rural/Urban']!='Rural'].index)
d10=d10.drop(d10[d10['Adolescent and youth']!='All Ages'].index)
d10=d10.drop([6])
d10.reset_index(drop=True,inplace=True)
d10['TotalSC']=d10['LiteratesSC-M']+d10['LiteratesSC-F']+d10['IlliteratesSC-M']+d10['IlliteratesSC-F']
d10['Total LiteratesSC']=d10['LiteratesSC-M']+d10['LiteratesSC-F']
d10['SC Literacy Rate']=(d10['Total LiteratesSC']*100)/d10['TotalSC']
d10=d10[['Area Name','SC Literacy Rate']]
d10=d10.dropna()
d10
```

# Q11. Tell whether in youth ages the no. of marginal workers is more or the no. of main workers.

```python
d11=df  
d11=d11.drop(d11[d11['Total/Rural/Urban']!='Total'].index)
d11=d11.drop(d11[d11['Adolescent and youth']!='Youth (15-24)'].index)
d11=d11.reset_index(drop=True)
d11=d11.drop([0])
d11=d11.reset_index(drop=True)
d11['TotalMarginal']=d11['MarginalWorker-M']+d11['MarginalWorker-F']
d11['TotalMainWorker']=d11['MainWorker-F']+d11['MainWorker-M']
TotalMarginal = d11['TotalMarginal'].sum()
TotalMain = d11['TotalMainWorker'].sum()
if(TotalMarginal>TotalMain):
    print("In youth ages the no. of marginal workers is more by {}".format(TotalMarginal-TotalMain))
else:
    print("In youth ages the no. of main workers is more by {}".format(TotalMain-TotalMarginal))
```

# Q13. Which state is home to the maximum no. of ST in India? Which state has the minimum no. of ST in India?


```python
d13=df  
d13=d13.drop(d13[d13['Total/Rural/Urban']!='Total'].index)
d13=d13.drop(d13[d13['Adolescent and youth']!='All Ages'].index)
d13=d13.reset_index(drop=True)
d13=d13.drop([0])
d13=d13.reset_index(drop=True)
d13['Total ST']=d13['ST-M']+d13['ST-F']
d13['Total']=d13['Total Male']+d13['Total Female']

d13['Percent'] = (d13['Total ST']*100) / d13['Total']
d113=d13[['Area Name','Percent']]
print("State Which is home to maximum no. of people belonging to Scheduled Tribes in India is ",end='\n\n')
print(d113[d113['Percent']==d113['Percent'].max()],end='\n\n')
print("State Which is home to minimum no. of people belonging to Scheduled Tribes in India is ",end='\n\n')
print(d113[d113['Percent']==d113['Percent'].min()])
```

# Q15. Out of total ST population in India, how much percentage of ST live in urban areas? And what perccentage of them are illiterate?

```python
a=(d15['Total ST Population'].values[2]*100)/d15['Total ST Population'].values[0]
b=a=(d15['Total Illiterate ST'].values[2]*100)/d15['Total ST Population'].values[2]
print("Percentage of people belonging to Scheduled Tribes living in urban areas is {} %".format(a))
print("Percentage of people belonging to Scheduled Tribes living in urban areas who are illiterate is {} %".format(b))
```
