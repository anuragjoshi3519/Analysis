# Q1. Which state in India has the maximum balanced male to female ratio?

```python
d1=df  
d1=d1.drop(d1[d1['Total/Rural/Urban']!='Total'].index)
d1=d1.drop(d1[d1['Adolescent and youth']!='All Ages'].index)
d1=d1.reset_index(drop=True)
d1=d1.drop([0])
```

```python
d1['Sex Ratio']=d1['Total Male']/d1['Total Female']
d=d1[['Area Name','Total Male','Total Female','Sex Ratio']]
d
```

```python
d[d['Sex Ratio']==d['Sex Ratio'].max()]
```
